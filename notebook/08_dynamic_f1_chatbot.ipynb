```python
import pandas as pd
import google.generativeai as genai
import os

from dotenv import load_dotenv
```

```python
load_dotenv()

GEMINI_API_KEY = os.getenv(
    "GEMINI_API_KEY"
)

genai.configure(
    api_key=GEMINI_API_KEY
)
```

```python
model = genai.GenerativeModel(
    "gemini-2.5-flash"
)
```

```python
drivers = pd.read_csv(
    "../data/drivers.csv"
)

constructors = pd.read_csv(
    "../data/constructors.csv"
)

races = pd.read_csv(
    "../data/races.csv"
)

results = pd.read_csv(
    "../data/results.csv"
)
```

```python
def get_dataset_summary():

    return f"""
    Drivers: {len(drivers)}
    Constructors: {len(constructors)}
    Races: {len(races)}
    Results: {len(results)}
    """
```

```python
print(
    get_dataset_summary()
)
```

```python
user_question = input(
    "Ask an F1 question: "
)
```

```python
prompt = f"""
You are an F1 Data Assistant.

Dataset Summary:

{get_dataset_summary()}

User Question:

{user_question}

Answer clearly and professionally.
"""
```

```python
response = model.generate_content(
    prompt
)

print(
    response.text
)
```

```python
while True:

    user_question = input(
        "\nAsk an F1 question (type 'exit' to quit): "
    )

    if user_question.lower() == "exit":
        print("Goodbye!")
        break

    prompt = f"""
    You are an F1 Data Assistant.

    Dataset Summary:

    {get_dataset_summary()}

    User Question:

    {user_question}

    Answer clearly and professionally.
    """

    response = model.generate_content(
        prompt
    )

    print("\nAnswer:")
    print(response.text)
```

```python
print(
    "Dynamic chatbot working successfully."
)
```
