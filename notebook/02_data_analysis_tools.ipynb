```python
import pandas as pd
```

```python
results = pd.read_csv("../data/results.csv")
drivers = pd.read_csv("../data/drivers.csv")
constructors = pd.read_csv("../data/constructors.csv")
```

```python
drivers["full_name"] = (
    drivers["forename"] +
    " " +
    drivers["surname"]
)

drivers.head()
```

```python
def top_drivers_by_wins():
    
    winners = results[
        results["positionOrder"] == 1
    ]
    
    win_counts = (
        winners["driverId"]
        .value_counts()
        .head(10)
        .reset_index()
    )
    
    win_counts.columns = [
        "driverId",
        "wins"
    ]
    
    return win_counts.merge(
        drivers[
            ["driverId", "full_name"]
        ],
        on="driverId"
    )
```

```python
top_drivers_by_wins()
```

```python
def top_constructors():
    
    winners = results[
        results["positionOrder"] == 1
    ]
    
    wins = (
        winners["constructorId"]
        .value_counts()
        .head(10)
        .reset_index()
    )
    
    wins.columns = [
        "constructorId",
        "wins"
    ]
    
    return wins.merge(
        constructors[
            ["constructorId", "name"]
        ],
        on="constructorId"
    )
```

```python
top_constructors()
```

```python
def driver_stats(driver_id):
    
    driver_results = results[
        results["driverId"] == driver_id
    ]
    
    return {
        "races":
            len(driver_results),
        
        "wins":
            len(
                driver_results[
                    driver_results[
                        "positionOrder"
                    ] == 1
                ]
            ),
        
        "avg_points":
            round(
                driver_results[
                    "points"
                ].mean(),
                2
            )
    }
```

```python
driver_stats(1)
```
