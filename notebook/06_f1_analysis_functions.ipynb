```python
import pandas as pd
```

```python
drivers = pd.read_csv("../data/drivers.csv")
constructors = pd.read_csv("../data/constructors.csv")
races = pd.read_csv("../data/races.csv")
results = pd.read_csv("../data/results.csv")
```

```python
def get_total_drivers():
    return len(drivers)

print(
    get_total_drivers()
)
```

```python
def get_total_races():
    return len(races)

print(
    get_total_races()
)
```

```python
def get_total_constructors():
    return len(constructors)

print(
    get_total_constructors()
)
```

```python
def get_total_results():
    return len(results)

print(
    get_total_results()
)
```

```python
def search_driver(name):

    filtered = drivers[
        drivers["surname"]
        .str.contains(
            name,
            case=False,
            na=False
        )
    ]

    return filtered[
        [
            "forename",
            "surname",
            "nationality"
        ]
    ]
```

```python
search_driver("hamilton")
```

```python
def search_constructor(name):

    filtered = constructors[
        constructors["name"]
        .str.contains(
            name,
            case=False,
            na=False
        )
    ]

    return filtered[
        [
            "name",
            "nationality"
        ]
    ]
```

```python
search_constructor("Ferrari")
```

```python
def search_race(name):

    filtered = races[
        races["name"]
        .str.contains(
            name,
            case=False,
            na=False
        )
    ]

    return filtered[
        [
            "year",
            "name",
            "date"
        ]
    ]
```

```python
search_race("Australian")
```
