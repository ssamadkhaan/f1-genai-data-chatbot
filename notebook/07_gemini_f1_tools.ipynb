```python
import pandas as pd
import google.generativeai as genai
import os

from dotenv import load_dotenv
```

```python
load_dotenv()

GEMINI_API_KEY = os.getenv(
    "GEMINI_API_KEY"
)

genai.configure(
    api_key=GEMINI_API_KEY
)
```

```python
model = genai.GenerativeModel(
    "gemini-2.5-flash"
)
```

```python
drivers = pd.read_csv(
    "../data/drivers.csv"
)

constructors = pd.read_csv(
    "../data/constructors.csv"
)

races = pd.read_csv(
    "../data/races.csv"
)

results = pd.read_csv(
    "../data/results.csv"
)
```

```python
def get_total_drivers():
    return len(drivers)


def get_total_constructors():
    return len(constructors)


def get_total_races():
    return len(races)


def get_total_results():
    return len(results)
```

```python
print(
    "Drivers:",
    get_total_drivers()
)

print(
    "Constructors:",
    get_total_constructors()
)

print(
    "Races:",
    get_total_races()
)

print(
    "Results:",
    get_total_results()
)
```

```python
question = """
How many drivers are available
in the dataset?
"""
```

```python
dataset_answer = f"""
The F1 dataset contains
{get_total_drivers()} drivers.
"""
```

```python
prompt = f"""
User Question:

{question}

Dataset Information:

{dataset_answer}

Answer the question
using the dataset information.
"""
```

```python
response = model.generate_content(
    prompt
)

print(
    response.text
)
```

```python
question = """
How many races are available
in the dataset?
"""
```

```python
dataset_answer = f"""
The dataset contains
{get_total_races()} races.
"""
```

```python
prompt = f"""
User Question:

{question}

Dataset Information:

{dataset_answer}

Answer the question.
"""
```

```python
response = model.generate_content(
    prompt
)

print(
    response.text
)
```

```python
print(
    "Gemini successfully used F1 dataset information."
)

print(
    "Tool integration is working."
)

print(
    "Ready for advanced chatbot development."
)
```
