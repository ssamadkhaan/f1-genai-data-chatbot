```python
import google.generativeai as genai

from dotenv import load_dotenv

import os
```

```python
load_dotenv()

GEMINI_API_KEY = os.getenv(
    "GEMINI_API_KEY"
)

print(
    GEMINI_API_KEY[:10]
)
```

```python
genai.configure(
    api_key=GEMINI_API_KEY
)
```

```python
model = genai.GenerativeModel(
    "gemini-2.5-flash"
)
```

```python
response = model.generate_content(
    "Say hello"
)

print(
    response.text
)
```

```python
response = model.generate_content(
    "What is Formula 1?"
)

print(response.text)
```
