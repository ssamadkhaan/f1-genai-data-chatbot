```python
import pandas as pd
import google.generativeai as genai
import os

from dotenv import load_dotenv
```

```python
load_dotenv()

GEMINI_API_KEY = os.getenv(
    "GEMINI_API_KEY"
)

genai.configure(
    api_key=GEMINI_API_KEY
)
```

```python
model = genai.GenerativeModel(
    "gemini-2.5-flash"
)
```

```python
drivers = pd.read_csv(
    "../data/drivers.csv"
)

constructors = pd.read_csv(
    "../data/constructors.csv"
)

races = pd.read_csv(
    "../data/races.csv"
)

results = pd.read_csv(
    "../data/results.csv"
)
```

```python
def search_driver(name):

    filtered = drivers[
        drivers["surname"]
        .str.contains(
            name,
            case=False,
            na=False
        )
    ]

    return filtered[
        [
            "forename",
            "surname",
            "nationality"
        ]
    ]
```

```python
def search_constructor(name):

    filtered = constructors[
        constructors["name"]
        .str.contains(
            name,
            case=False,
            na=False
        )
    ]

    return filtered[
        [
            "name",
            "nationality"
        ]
    ]
```

```python
def search_race(name):

    filtered = races[
        races["name"]
        .str.contains(
            name,
            case=False,
            na=False
        )
    ]

    return filtered[
        [
            "year",
            "name",
            "date"
        ]
    ]
```

```python
print(
    search_driver(
        "hamilton"
    )
)
```

```python
print(
    search_constructor(
        "Ferrari"
    )
)
```

```python
print(
    search_race(
        "Australian"
    )
)
```

```python
while True:

    user_question = input(
        "\nAsk an F1 question (exit to quit): "
    )

    if user_question.lower() == "exit":
        print("Goodbye!")
        break

    tool_data = ""

    if "hamilton" in user_question.lower():

        tool_data = search_driver(
            "hamilton"
        ).to_string()

    elif "ferrari" in user_question.lower():

        tool_data = search_constructor(
            "Ferrari"
        ).to_string()

    elif "australian" in user_question.lower():

        tool_data = search_race(
            "Australian"
        ).to_string()

    prompt = f"""
    You are an F1 Data Assistant.

    User Question:
    {user_question}

    Tool Data:
    {tool_data}

    Answer using the tool data.
    """

    response = model.generate_content(
        prompt
    )

    print("\nAnswer:")
    print(response.text)
```

```python
print(
    "Smart F1 chatbot ready."
)
```
