```python
import pandas as pd
import google.generativeai as genai
import os

from dotenv import load_dotenv

load_dotenv()

GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")

genai.configure(
    api_key=GEMINI_API_KEY
)

model = genai.GenerativeModel(
    "gemini-2.5-flash"
)

print("Gemini connected.")
```

```python
drivers = pd.read_csv("../data/drivers.csv")
constructors = pd.read_csv("../data/constructors.csv")
races = pd.read_csv("../data/races.csv")
results = pd.read_csv("../data/results.csv")

print("Data loaded.")
```

```python
def get_driver_info(name):

    matches = drivers[
        drivers["surname"].str.contains(
            name,
            case=False,
            na=False
        )
    ]

    return matches.head(10)
```

```python
def get_constructor_info(name):

    matches = constructors[
        constructors["name"].str.contains(
            name,
            case=False,
            na=False
        )
    ]

    return matches.head(10)
```

```python
def get_race_info(name):

    matches = races[
        races["name"].str.contains(
            name,
            case=False,
            na=False
        )
    ]

    return matches.head(20)
```

```python
def select_tool(question):

    prompt = f"""
You are a routing assistant.

Choose only ONE word.

DRIVER
CONSTRUCTOR
RACE

Question:
{question}
"""

    response = model.generate_content(prompt)

    return response.text.strip().upper()
```

```python
print(
    select_tool(
        "Tell me about Hamilton"
    )
)
```

```python
print(
    select_tool(
        "Tell me about Ferrari"
    )
)
```

```python
print(
    select_tool(
        "Show Australian Grand Prix races"
    )
)
```

```python
while True:

    question = input(
        "Ask an F1 question (exit to quit): "
    )

    if question.lower() == "exit":
        break

    tool = select_tool(question)

    print("Selected Tool:", tool)

    if tool == "DRIVER":

        keyword = question.split()[-1]

        data = get_driver_info(keyword)

    elif tool == "CONSTRUCTOR":

        keyword = question.split()[-1]

        data = get_constructor_info(keyword)

    else:

        keyword = question.replace(
            "Show",
            ""
        ).replace(
            "races",
            ""
        )

        data = get_race_info(keyword)

    prompt = f"""
Question:
{question}

Relevant Data:
{data}

Answer professionally.
"""

    response = model.generate_content(
        prompt
    )

    print("\nAnswer:")
    print(response.text)
    print("\n")
```

```python
print(
    "Auto tool selector ready."
)
```
