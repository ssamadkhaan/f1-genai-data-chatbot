```python
import pandas as pd
import google.generativeai as genai
import os

from dotenv import load_dotenv

load_dotenv()

GEMINI_API_KEY = os.getenv(
    "GEMINI_API_KEY"
)

genai.configure(
    api_key=GEMINI_API_KEY
)

model = genai.GenerativeModel(
    "gemini-2.5-flash"
)

print("Gemini connected.")
```

```python
drivers = pd.read_csv(
    "../data/drivers.csv"
)

constructors = pd.read_csv(
    "../data/constructors.csv"
)

races = pd.read_csv(
    "../data/races.csv"
)

results = pd.read_csv(
    "../data/results.csv"
)

print("Data loaded.")
```

```python
def get_driver_info(name):

    return drivers[
        drivers["surname"].str.contains(
            name,
            case=False,
            na=False
        )
    ].head(10)
```

```python
def get_constructor_info(name):

    return constructors[
        constructors["name"].str.contains(
            name,
            case=False,
            na=False
        )
    ].head(10)
```

```python
def get_race_info(name):

    return races[
        races["name"].str.contains(
            name,
            case=False,
            na=False
        )
    ].head(20)
```

```python
def analyze_question(question):

    prompt = f"""
You are an F1 assistant.

Return only:

TOOL=<DRIVER/CONSTRUCTOR/RACE>
KEYWORD=<keyword>

Question:
{question}
"""

    response = model.generate_content(
        prompt
    )

    return response.text
```

```python
result = analyze_question(
    "Tell me about Lewis Hamilton"
)

print(result)
```

```python
result = analyze_question(
    "Tell me about Mercedes"
)

print(result)
```

```python
def parse_response(text):

    lines = text.strip().split("\n")

    tool = lines[0].replace(
        "TOOL=",
        ""
    ).strip()

    keyword = lines[1].replace(
        "KEYWORD=",
        ""
    ).strip()

    return tool, keyword
```

```python
response = analyze_question(
    "Tell me about Hamilton"
)

tool, keyword = parse_response(
    response
)

print(tool)
print(keyword)
```

```python
while True:

    question = input(
        "Ask an F1 question (exit to quit): "
    )

    if question.lower() == "exit":
        break

    router_response = analyze_question(
        question
    )

    tool, keyword = parse_response(
        router_response
    )

    print(
        "\nTool:",
        tool
    )

    print(
        "Keyword:",
        keyword
    )

    if tool == "DRIVER":

        data = get_driver_info(
            keyword
        )

    elif tool == "CONSTRUCTOR":

        data = get_constructor_info(
            keyword
        )

    else:

        data = get_race_info(
            keyword
        )

    prompt = f"""
Question:
{question}

Relevant Data:
{data}

Answer professionally.
"""

    answer = model.generate_content(
        prompt
    )

    print("\nAnswer:")
    print(answer.text)
    print("\n")
```

```python
print(
    "AI tool and keyword selector ready."
)
```
