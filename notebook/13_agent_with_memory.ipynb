```python
import os
from dotenv import load_dotenv

import google.generativeai as genai
```

```python
load_dotenv()

GEMINI_API_KEY = os.getenv(
    "GEMINI_API_KEY"
)

genai.configure(
    api_key=GEMINI_API_KEY
)
```

```python
model = genai.GenerativeModel(
    "gemini-2.5-flash"
)
```

```python
chat = model.start_chat(
    history=[]
)
```

```python
response = chat.send_message(
    "Tell me about Lewis Hamilton"
)

print(
    response.text
)
```

```python
response = chat.send_message(
    "Which team does he drive for?"
)

print(
    response.text
)
```

```python
response = chat.send_message(
    "What nationality is he?"
)

print(
    response.text
)
```

```python
for message in chat.history:

    print(
        message.role
    )

    print(
        message.parts[0].text
    )

    print("-" * 50)
```

```python
while True:

    question = input(
        "Ask an F1 question (exit to quit): "
    )

    if question.lower() == "exit":
        break

    response = chat.send_message(
        question
    )

    print("\nAnswer:\n")

    print(
        response.text
    )

    print("\n")
```

```python
print(
    "Agent with memory ready."
)
```
