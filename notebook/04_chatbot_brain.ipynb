```python
import google.generativeai as genai
from dotenv import load_dotenv
import os
```

```python
load_dotenv()

GEMINI_API_KEY = os.getenv(
    "GEMINI_API_KEY"
)
```

```python
genai.configure(
    api_key=GEMINI_API_KEY
)
```

```python
model = genai.GenerativeModel(
    "gemini-2.5-flash"
)
```

```python
def ask_chatbot(question):

    response = model.generate_content(
        question
    )

    return response.text
```

```python
answer = ask_chatbot(
    "Who is Lewis Hamilton?"
)

print(answer)
```

```python
while True:

    question = input(
        "Ask: "
    )

    if question.lower() == "exit":
        break

    answer = ask_chatbot(
        question
    )

    print("\nBot:")
    print(answer)
    print()
```

## Conclusion

A reusable AI chatbot was successfully developed using the Gemini API. The chatbot can accept user questions dynamically and generate responses in real time.
