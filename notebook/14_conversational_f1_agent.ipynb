```python
import os
from dotenv import load_dotenv

import pandas as pd
import google.generativeai as genai
```

```python
load_dotenv()

GEMINI_API_KEY = os.getenv(
    "GEMINI_API_KEY"
)

genai.configure(
    api_key=GEMINI_API_KEY
)
```

```python
drivers = pd.read_csv(
    "../data/drivers.csv"
)

constructors = pd.read_csv(
    "../data/constructors.csv"
)
```

```python
model = genai.GenerativeModel(
    "gemini-2.5-flash"
)
```

```python
chat = model.start_chat(
    history=[]
)
```

```python
def get_driver_info(name):

    result = drivers[
        (
            drivers["forename"]
            .fillna("")
            .str.contains(
                name,
                case=False,
                na=False
            )
        )
        |
        (
            drivers["surname"]
            .fillna("")
            .str.contains(
                name,
                case=False,
                na=False
            )
        )
    ]

    return result.head(5)
```

```python
def get_constructor_info(name):

    result = constructors[
        constructors["name"]
        .str.contains(
            name,
            case=False,
            na=False
        )
    ]

    return result.head(5)
```

```python
while True:

    question = input(
        "Ask an F1 question (exit to quit): "
    )

    if question.lower() == "exit":
        break

    driver_data = get_driver_info(
        question
    )

    constructor_data = get_constructor_info(
        question
    )

    prompt = f"""
User Question:
{question}

Driver Data:
{driver_data}

Constructor Data:
{constructor_data}

Use the provided data if available.
If no matching data exists, answer using your knowledge.
"""

    response = chat.send_message(
        prompt
    )

    print("\nAnswer:\n")

    print(
        response.text
    )

    print("\n")
```

```python
print(
    "Conversational F1 Agent ready."
)
```
